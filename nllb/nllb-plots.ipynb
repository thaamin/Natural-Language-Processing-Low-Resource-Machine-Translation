```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define the data
char_scores = [0.431, 0.550, 0.579, 0.579, 0.571, 0.560]
word_scores = [0.537, 0.531, 0.481, 0.414]

# Calculate averages for annotation
avg_char = sum(char_scores) / len(char_scores)
avg_word = sum(word_scores) / len(word_scores)

# 2. Prepare Data for Plotting
df = pd.DataFrame({
    'Metric Type': (['Character-level'] * len(char_scores)) + (['Word-level'] * len(word_scores)),
    'Pearson Correlation (r)': char_scores + word_scores
})

# 3. Create the Boxplot
plt.style.use('ggplot')
plt.figure(figsize=(8, 6))

# Boxplot with a white dot for the mean
ax = sns.boxplot(
    x='Metric Type',
    y='Pearson Correlation (r)',
    data=df,
    palette=['#2E86C1', '#E67E22'],
    width=0.4,
    showmeans=True,
    meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"black", "markersize":"10"}
)

# Overlay points to show the specific n-gram distributions
sns.stripplot(x='Metric Type', y='Pearson Correlation (r)', data=df, color='black', alpha=0.3)

# 4. Text Annotations
plt.title('Metric Reliability: Character vs. Word (Slavic Proxy)', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Correlation with Human Judgment', fontsize=12)
plt.xlabel('')

# Add average value labels above the boxes
plt.text(0, avg_char + 0.015, f'Mean: {avg_char:.3f}', ha='center', color='#2E86C1', weight='bold')
plt.text(1, avg_word + 0.015, f'Mean: {avg_word:.3f}', ha='center', color='#E67E22', weight='bold')

plt.tight_layout()
plt.savefig('metric_comparison_boxplot.png', dpi=300)
```

```python
import pandas as pd
from plotnine import ggplot, aes, geom_bar
df_nllb =pd.DataFrame({"Model": ["Base NLLB", "Fully-Trained NLLB"],"Bleu": [6, 26.01],"Chrf++":[23, 63.09363116005535]})
df_qwen =pd.DataFrame({"Model": ["Base Qwen", "few-shot Qwen" ,"Fully-Trained Qwen"],"Bleu": [0.06 , 0.01, 53.5],"Chrf++":[4.24, 1.23 ,73.74 ]})
```

```python
from plotnine import ggplot, aes, geom_col, labs, theme_minimal, ylim

# BLEU scores for NLLB
p_nllb = (
    ggplot(df_nllb, aes(x='Model', y='Bleu', fill='Model')) +  # fill for color
    geom_col() +  # bar plot
    ylim(0, 100) +
    labs(title="BLEU Scores of NLLB Models", x="Model", y="BLEU Score") +
    theme_minimal()
)

print(p_nllb.show())
```

```python
order = ["Base Qwen",  "few-shot Qwen" ,"Fully-Trained Qwen"]
df_qwen["Model"] = pd.Categorical(df_qwen["Model"], categories=order ,ordered = True)
p_nllb = (
    ggplot(df_qwen, aes(x='Model', y='Bleu', fill='Model')) +  # fill for color
    geom_col() +  # bar plot
    ylim(0, 100) +
    theme_minimal()
)

print(p_nllb.show())
```
